```python
plt.Figure(figsize=(6,6))
title_size = 21

base_color = '#00ccff'
name = ['Benign', 'malignant']
p = [base_color, 'black']
plt.pie(df.target.value_counts(normalize=True), autopct='%1.2f%%', colors = sns.color_palette(p),
 startangle=90, wedgeprops=dict(width=0.25), labeldistance=1.2, counterclock=False, radius=1)
plt.title(f'Target distribution', color = 'black', fontsize=title_size)
plt.legend(name,loc='center right', prop={'size':12},markerscale =3, framealpha=0.8, facecolor='white')
plt.show()
```
